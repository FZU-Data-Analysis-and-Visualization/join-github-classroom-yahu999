### 四、分组聚合

1. 按 gender 分组，求 score 的均值和最大值。
2. 按 gender 分组，统计每组人数。

```python
data = {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
aa=pd.DataFrame(data)
aa
```

```python
#1
aa.groupby('gender')['score'].agg(['mean','max'])
```

```python
#2
aa.groupby('gender').size()
```

### 六、合并与拼接

1. 横向合并：

   df1 = pd.DataFrame ({"id":[1,2], "name":["X","Y"]})
   df2 = pd.DataFrame ({"id":[1,2], "age":[20,30]})

   按 id 合并。   

2. 纵向拼接：

   df_a = pd.DataFrame ({"val":[1,2]})

   df_b = pd.DataFrame ({"val":[3,4]})

   拼接后重置索引。

```python
#1
df1 = pd.DataFrame({"id": [1, 2], "name": ["X", "Y"]})
df2 = pd.DataFrame({"id": [1, 2], "age": [20, 30]})
print(df1)
print(df2)
```

```python
#横向合并
df3= pd.merge(df1, df2, on="id")
df3
```

```python
#2
df_a = pd.DataFrame ({"val":[1,2]})
df_b = pd.DataFrame ({"val":[3,4]})

df4= pd.concat([df_a, df_b], ignore_index=True)
df4
```

### 七、时间处理

1. 创建时间列：["2025-01-01","2025-01-02","2025-01-03"]，转为 datetime 格式。
2. 提取时间列的月份、星期。

```python
#1
time=pd.DataFrame({"date": ["2025-01-01", "2025-01-02", "2025-01-03"]})
time['date'] = pd.to_datetime(time['date'])
time
```

```python
#2
time['month'] = time['date'].dt.month
time['weekday'] = time['date'].dt.weekday
time
```

### 八、高级操作

1. 对 score 列做标准化（(值 - 均值)/ 标准差）。

```python
#1
bb['score_std'] = (bb['score'] - bb['score'].mean()) / bb['score'].std()
bb
```
